```python
from scipy.integrate import odeint
import numpy as np
from math import cos, sin
import scipy
import matplotlib.pyplot as plt

# Simulation parameters
g = 9.81
m = 0.5
J = np.diag([0.0557, 0.0557, 0.105]) # moment of Inertia in Kg(m^2)

# Interial fram axis
e1 = np.array([[1,0,0]]).T
e2 = np.array([[0,1,0]]).T
e3 = np.array([[0,0,1]]).T
d = 0.315    # distance of center of mass from fram center in m
c = 8.004*10e-4 # fixed constant in m
```

```python
def hat_map(v):

    return [[0, -v[2][0], v[1][0]], [v[2][0], 0 ,-v[0][0]], [-v[1][0], v[0][0], 0]]

def vee_map(R):
    return np.array([[R[2, 1],R[0, 2],R[1, 0]]])
def vec_cross(a, b):
    tmp = np.array([[(a[1]*b[2]-a[2]*b[1]),(-a[0]*b[2] + a[2]*b[0]),(a[0]*b[1] - a[1]*b[0])]]).T
    ans = np.squeeze(tmp, axis=0)
    return ans
```

```python
# ode function 
def step(y,t):
    x_curr = np.array([y[0:3]]).T
    v_curr = np.array([y[3:6]]).T
    R = y[6:15].reshape((3,3))
    w = np.array([y[15:18]]).T
    b3 =np.array([ R[:,2] ]).T 
    
    # parameters
    k1 = np.diag([5, 5 ,9])
    k2 = np.diag([5, 5 ,10])

    # case 1 parameters
    kR = 200
    kOm = 1 

    # desired states
    des_pos = np.array([[0.4*t, 0.4*sin(np.pi*t), 0.6*cos(np.pi*t)]]).T
    des_vel = np.array([[0.4, 0.4* np.pi* cos(np.pi*t), -0.6*np.pi*sin(np.pi*t)]]).T
    des_acc = np.array([[0, -0.4*np.pi*np.pi*sin(np.pi*t),-0.6*np.pi*np.pi*cos(np.pi*t)]]).T 

    b1d = np.array([[cos(np.pi*t), sin(np.pi*t), 0]]).T
    w_desired = np.array([[0,0,0]]).T

    # Position Control
    pos_e = np.subtract(x_curr, des_pos)
    vel_e = np.subtract(v_curr, des_vel)
    A = (-np.dot(k1,pos_e) -np.dot(k2,vel_e) + np.dot(m,des_acc) + np.dot(m*g,e3))

    b3_desired = A/np.linalg.norm(A) 
 
    f = np.dot(A.T,b3)
    

    
    # Attitude Control
    print('b3d',b3_desired)
    b2d = np.cross(b3_desired,b1d)
    b2d = b2d/np.linalg.norm(b2d) 

    projection_b1d = -np.cross(b3_desired,b2d)
    temp = np.cross(b2d,b3_desired)
#     Rd = np.array([np.cross(b2d,b3_desired), b2d,  b3_desired])
    Rd =np.concatenate((temp, b2d,  b3_desired.T),axis = 0)
    Rd = Rd.T


#     projection_b1d = -vec_cross(b3_desired,b2d)
#     check = vec_cross(b2d,b3_desired)
#     Rd_middle = np.concatenate((check, b2d), axis=1)
#     Rd = np.concatenate((Rd_middle, b3_desired), axis=1)

    
    # calculating error in angular velocity and attitude
    R_e = 1/2 * vee_map(np.subtract(np.dot(Rd.T, R), np.dot(R.T, Rd)))
    R_e = R_e[..., np.newaxis]
    
    
    
    Om_e = w - np.dot(R.T, (np.dot(Rd,w_desired)))

    B = np.dot(R.T, np.dot(Rd, w_desired))

    C = np.dot(hat_map(w), B) - B
#     print('c',np.cross(w.T, np.dot(J,w).T))
#     print(-np.dot(kR,R_e.T))
#     print(- np.dot(kOm,Om_e))
#     print(np.cross(w.T, np.dot(J,w).T).T)
#     print(np.dot(J,C))
    M = -(kR * R_e) - (kOm * Om_e) + np.cross(w, np.dot(J,w)) - np.dot(J,C)
#     print(-np.dot(g,e3))
#     print(np.dot((f/m)*R, e3.T).T)
    v_dot = np.dot(-g,e3) + np.dot((f/m)*R, e3)

    R_dot = np.dot(R, hat_map(w))    

#     Omega_dot = (-np.cross(w.T, np.dot(J,w).T).T + M )
    D = (-np.cross(w, np.dot(J,w)) + M )
    Omega_dot = np.dot(np.linalg.inv(J),D)
#     print(Omega_dot[0])
#     w1 = Omega_dot[0][0] / np.diag(J)[0]
#     w2 = Omega_dot[1][0] / np.diag(J)[1]
#     w3 = Omega_dot[2][0] / np.diag(J)[2]
#     print(v_curr[0])
#     print(v_dot[0])
#     print(R_dot.reshape((1,9))[0])
#     print(Omega_dot.T[0])
    dydt = np.concatenate((v_curr, v_dot, R_dot.reshape((9,1)),Omega_dot), axis=0)
    dydt = np.squeeze(dydt, axis=1)
#     print(dydt)
    return dydt
```

```python
R0 = np.array([1, 0, 0, 0, 1, 0, 0, 0, 1])
y0 = np.concatenate((0, 0, 0.0, 0, 0, 0, R0,0, 0, 0,), axis=None)
```

```python
t = np.linspace(0, 20, 100)
sol = odeint(step, y0, t)
```

```python
# plot solution 
plt.plot(t, sol[:, 0], 'b', label='x')
plt.plot(t, sol[:, 1], 'g', label='y')
plt.plot(t, sol[:, 2], 'r', label='z')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
# Compare desired pos with sol 
x = []
y = []
z = [] 

for i in range(len(t)):
#     x.append(0)
    x.append(0.4*t[i])
    y.append(0.4*sin(np.pi*t[i]))
    z.append(0.6*cos(np.pi*t[i]))
x.append(0.4*t[49]) 
# x.append(0)
y.append(0.4*sin(np.pi*t[49])) 
z.append(0.6*cos(np.pi*t[49])) 
plt.plot(t, x[1:], label='x_d')
plt.plot(t, sol[:, 0], 'b', label='x')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()

plt.plot(t, y[1:], label='y_d')
plt.plot(t, sol[:, 1], 'g', label='y')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()

plt.plot(t, z[1:], label='z_d')
plt.plot(t, sol[:, 2], 'r', label='z')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

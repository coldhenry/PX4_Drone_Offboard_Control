```python
from scipy.integrate import odeint
import numpy as np
from math import cos, sin
import scipy
import matplotlib.pyplot as plt

# Simulation parameters
g = 9.81
m = 0.5
J = np.diag([0.0557, 0.0557, 0.105]) # moment of Inertia in Kg(m^2)

# Interial fram axis
e1 = np.array([1,0,0])
e2 = np.array([0,1,0])
e3 = np.array([0,0,1])
d = 0.315    # distance of center of mass from fram center in m
c = 8.004*10e-4 # fixed constant in m
```

```python
def hat_map(v):

    return [[0, -v[2], v[1]], [v[2], 0 ,-v[0]], [-v[1], v[0], 0]]

def vee_map(R):
    return np.array([R[2, 1],R[0, 2],R[1, 0]])
```

```python
# ode function 
def step(y,t):
    x_curr = y[0:3]
    v_curr = y[3:6]
    R = y[6:15].reshape((3,3))
    w = y[15:18]
    b3 = R[:,2]  
    
    # parameters
    k1 = np.diag([5, 5 ,9])
    k2 = np.diag([5, 5 ,10])
#     k1 = np.diag([5, 5 ,50])
#     k2 = np.diag([5, 5 ,9])

    # case 1 parameters
    kR = 200
    kOm = 1 

    # desired states
    des_pos = np.array([0.4*t, 0.4*sin(np.pi*t), 0.6*cos(np.pi*t)])
    des_vel = np.array([0.4, 0.4* np.pi* cos(np.pi*t), -0.6*np.pi*sin(np.pi*t)])
#     des_pos = np.array([0, 0.4*sin(np.pi*t), 0.6*cos(np.pi*t)])
#     des_vel = np.array([0, 0.4* np.pi* cos(np.pi*t), -0.6*np.pi*sin(np.pi*t)])
    des_acc = np.array([0, -0.4*np.pi*np.pi*sin(np.pi*t),-0.6*np.pi*np.pi*cos(np.pi*t)])  

#     b1d = np.array([cos(np.pi*t), sin(np.pi*t), 0])
    b1d = np.array([cos(np.pi*t), sin(np.pi*t), 0])
    w_desired = np.array([0,0,0])

    # Position Control
    pos_e = np.subtract(x_curr, des_pos)
    vel_e = np.subtract(v_curr, des_vel)
    A = (-np.dot(k1,pos_e) -np.dot(k2,vel_e) + np.dot(m,des_acc) + np.dot(m*g,e3))

    b3_desired = A/np.linalg.norm(A) 
    f = np.dot(A.T,b3)

    
    # Attitude Control

    b2d = np.cross(b3_desired,b1d)
    b2d = b2d/np.linalg.norm(b2d) 
    projection_b1d = -np.cross(b3_desired,b2d)
    temp = np.cross(b2d,b3_desired)/np.linalg.norm(np.cross(b2d,b3_desired))
#     Rd = np.array([np.cross(b2d,b3_desired), b2d,  b3_desired])
    Rd = np.array([temp, b2d,  b3_desired])

    # calculating error in angular velocity and attitude
    R_e = 1/2 * vee_map(np.subtract(np.dot(Rd.T, R), np.dot(R.T, Rd)))

    Om_e = w - np.dot(R.T, (np.dot(Rd,w_desired.T)))

    B = np.dot(R.T, np.dot(Rd, w_desired.T))

    C = np.dot(hat_map(w), B) - B

    M = -np.dot(kR,R_e) - np.dot(kOm,Om_e) + np.cross(w, np.dot(J,w)) - np.dot(J,C)

#     v_dot = -np.dot(g,e3) + np.dot((f/m)*R, e3) 
    v_dot = -np.dot(g,e3) + np.dot(np.dot((f/m),R), e3)
    
#     print(v_dot)

    R_dot = np.dot(R, hat_map(w))    
    
    D = (-np.cross(w, np.dot(J,w)) + M )
    
    Omega_dot = np.dot(np.linalg.inv(J),D)
#     print(Omega_dot)
#     w1 = (-np.cross(w, np.dot(J,w)) + M )[0] / np.diag(J)[0]
#     print('w1',w1)
    
#     w2 = Omega_dot[1] / np.diag(J)[1]
#     w3 = Omega_dot[2] / np.diag(J)[2]

    dydt = np.concatenate((v_curr, v_dot, R_dot.reshape((1,9)),Omega_dot), axis=None)
    return dydt
```

```python
R0 = np.array([1, 0, 0, 0, 1, 0, 0, 0, 1])
y0 = np.concatenate((0, 0, 0, 0, 0, 0, R0,0, 0, 0,), axis=None)
```

```python
t = np.linspace(0, 10, 100)
sol = odeint(step, y0, t)
```

```python
# plot solution 
plt.plot(t, sol[:, 0], 'b', label='x')
plt.plot(t, sol[:, 1], 'g', label='y')
plt.plot(t, sol[:, 2], 'r', label='z')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```

```python
# Compare desired pos with sol 
x = []
y = []
z = [] 

for i in range(len(t)):
#     x.append(0)
    x.append(0.4*t[i])
    y.append(0.4*sin(np.pi*t[i]))
    z.append(0.6*cos(np.pi*t[i]))
x.append(0.4*t[49]) 
# x.append(0)
y.append(0.4*sin(np.pi*t[49])) 
z.append(0.6*cos(np.pi*t[49])) 
plt.plot(t, x[1:], label='x_d')
plt.plot(t, sol[:, 0], 'b', label='x')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()

plt.plot(t, y[1:], label='y_d')
plt.plot(t, sol[:, 1], 'g', label='y')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()

plt.plot(t, z[1:], label='z_d')
plt.plot(t, sol[:, 2], 'r', label='z')
plt.legend(loc='best')
plt.xlabel('t')
plt.grid()
plt.show()
```
